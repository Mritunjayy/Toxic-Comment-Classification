```python
import re
def remove_punctuations_text(text):
    return re.sub(r'[^\w\s]', '', text)

# Apply the function to the 'text' column
train['comment_text'] = train['comment_text'].apply(remove_punctuations_text)
```

```python
train
```

```python
import re

def remove_special_characters(text):
    pattern = r'[^a-zA-Z0-9\s]'
    return re.sub(pattern, '', text)

train['comment_text'] = train['comment_text'].apply(remove_special_characters)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def remove_stopwords(sentence):
    tokens = sentence.split()
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    return ' '.join(filtered_tokens)

train['comment_text'] = train['comment_text'].apply(remove_stopwords)
```

```python
import pandas as pd
import spacy
from joblib import Parallel, delayed

# Load a smaller and faster spaCy model
nlp = spacy.load('en_core_web_sm', disable=['ner', 'parser'])

def lemmatize(sentence):
    return ' '.join([token.lemma_ for token in nlp(sentence)])

def batch_lemmatize(sentences):
    lemmatized_sentences = []
    for doc in nlp.pipe(sentences, batch_size=1000):
        lemmatized_sentences.append(' '.join([token.lemma_ for token in doc]))
    return lemmatized_sentences

train['comment_text_lemmatized'] = batch_lemmatize(train['comment_text'].tolist())
train
```

```python
from sklearn.model_selection import train_test_split

X = train['comment_text_lemmatized']
y = train.drop(columns=['comment_text', 'comment_text_lemmatized'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

X_tfidf = tfidf_vectorizer.fit_transform(X)

print("Shape of feature matrix (TF-IDF):", X_tfidf.shape)
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

# X_train: Training features (lemmatized comment text)
# X_test: Testing features (lemmatized comment text)
# y_train: Training labels
# y_test: Testing labels
print(X_train.shape)
print(X_test.shape)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.multiclass import OneVsRestClassifier

rf_estimator = RandomForestClassifier(n_estimators=21)

ovr_classifier = OneVsRestClassifier(rf_estimator)

ovr_classifier.fit(X_train, y_train)

y_prob = ovr_classifier.predict_proba(X_test)

auc_score = roc_auc_score(y_test, y_prob, average='macro')

print("Average AUC Score:", auc_score)
```

```python
from skmultilearn.problem_transform import BinaryRelevance
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score

nb_estimator = GaussianNB()

br_classifier = BinaryRelevance(classifier=nb_estimator)

br_classifier.fit(X_train, y_train)

y_prob = br_classifier.predict_proba(X_test)

auc_score = roc_auc_score(y_test, y_prob.toarray(), average='macro')
print("Average AUC Score:", auc_score)
```

```python
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

lr_estimator = LogisticRegression()

cc_classifier = ClassifierChain(classifier=lr_estimator)

cc_classifier.fit(X_train, y_train)

y_prob = cc_classifier.predict_proba(X_test)

auc_score = roc_auc_score(y_test, y_prob.toarray(), average='macro')

print("Average AUC Score:", auc_score)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score

knn_estimator = KNeighborsClassifier()

mo_classifier = MultiOutputClassifier(knn_estimator)

mo_classifier.fit(X_train, y_train)

y_prob = mo_classifier.predict_proba(X_test)

auc_score = roc_auc_score(y_test, y_prob, average='macro')

print("Average AUC Score:", auc_score)
```
